```python
import numpy as np 
import matplotlib.pyplot as plt
A = np.zeros((5,100))
i= len(A) 
for i in range(0,i) :
   var = 0.01 
   for j in range(1,100):
        A[i,j]+=var
        var +=0.01
print(A)
plt.imshow(A, cmap=plt.gray())
plt.imshow(A)
```

```python
import matplotlib.pyplot as plt
import numpy as np 

img = np.zeros((3,3,3))
R = np.array( [[0.5, 0, 1],      
               [0, 1, 0],          
               [1, 0, 0.5]]) 

 
G = np.array( [[0.5, 0.6, 1],       
               [0.6, 0, 0],       
               [0, 0.6, 0.5]])
B = np.array( [[0.5, 1, 0],          
               [1, 0, 1],       
               [0, 1, 0.5]])
img[:,:,0]=R 
img[:,:,1]=G 
img[:,:,2]=B 
plt.imshow(img, interpolation='nearest')
plt.axis('off')
plt.show() 
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
img = Image.open("brba.jpg")

i = 0
j = 0
jaune = (231,210,43)
for i in range(0,800) :
    for j in range(0,800):
        if (img.getpixel((i,j))) == jaune :
            
         img.putpixel((i,j),(0,255,0)) 
         
plt.imshow(img, interpolation='nearest')
plt.show()
    
```

```python
cap = cv2.VideoCapture(0) 
while True:     
    ret, frame = cap.read()   
    cv2.imshow("frame", frame) 
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray)
    key = cv2.waitKey(1)     
    if key == 27:         
        break 
cap.release() 
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture("red_panda_snow.mp4 ")
fourcc = cv2.VideoWriter_fourcc(*"XVID") 
out = cv2.VideoWriter("flipped_video.avi", fourcc, 50, (640, 360))
while True:    
    ret, frame = cap.read()   
    frame2 = cv2.flip(frame, 1)   
    cv2.imshow("frame2", frame2)   
    cv2.imshow("frame", frame)   
    out.write(frame2) 
```
